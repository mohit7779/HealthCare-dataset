#  ANOVA TEST

```python
import pandas as pd
import scipy.stats as stats

# Load the healthcare dataset
data = pd.read_csv(r"C:\Users\HP\Downloads\archive (10)\healthcare_dataset.csv")

# Check if the relevant columns are available
if 'Admission Type' in data.columns and 'Billing Amount' in data.columns:
    # Filter data for each admission type
    urgent_billing = data[data['Admission Type'] == 'Urgent']['Billing Amount']
    emergency_billing = data[data['Admission Type'] == 'Emergency']['Billing Amount']
    elective_billing = data[data['Admission Type'] == 'Elective']['Billing Amount']

    # Perform ANOVA test
    f_stat, p_value = stats.f_oneway(urgent_billing, emergency_billing, elective_billing)

    # Display results
    print("ANOVA Test Results:")
    print(f"F-statistic: {f_stat}")
    print(f"P-value: {p_value}")

    # Interpretation
    if p_value < 0.05:
        print("Result: Significant difference in mean billing amounts across admission types (p < 0.05).")
    else:
        print("Result: No significant difference in mean billing amounts across admission types (p >= 0.05).")
else:
    print("Required columns 'Admission_Type' and 'Billing_Amount' are not found in the dataset.")
```
